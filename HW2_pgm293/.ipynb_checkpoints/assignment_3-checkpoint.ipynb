```python
#Fill cells that do not havea Borough with the Text 'NO BOROUGH' to avoid indexing issues
mvcman = nycmvc[nycmvc['BOROUGH'].fillna('NO BOROUGH').str.contains("MANHATTAN")]
mvcman.head()
```

```python
mvcman.drop(['DATE', 'TIME', 'BOROUGH','LATITUDE','LONGITUDE',
       'LOCATION', 'ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME',
        'NUMBER OF PEDESTRIANS INJURED','NUMBER OF PEDESTRIANS KILLED',
       'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
       'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED',
       'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2',
       'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4',
       'CONTRIBUTING FACTOR VEHICLE 5', 'UNIQUE KEY', 'VEHICLE TYPE CODE 1',
       'VEHICLE TYPE CODE 2', 'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4',
       'VEHICLE TYPE CODE 5'],axis=1,inplace=True)
mvcman.head()
```

```python
ax = mvcman.plot(x='ZIP CODE', y='NUMBER OF PERSONS INJURED', kind="scatter",alpha=0.2,label="Zip Code vs People Injured")
ax.set_title("Motor Vehicle Collisions Manhattan")
#Tried adding some text but my kernel keeps crashing
```

```python
ax = mvcman.plot(x='ZIP CODE', y='NUMBER OF PERSONS KILLED', kind="scatter",alpha=0.2,label="Zip Code vs Number of Persons Killed")
ax.set_title("Motor Vehicle Collisions Manhattan")
```

```python
ax = mvcman.plot(x='ZIP CODE', y='NUMBER OF PERSONS INJURED', kind="scatter",alpha=0.2,color="SteelBlue")
ax.set_title("Motor Vehicle Collisions Manhattan")

mvcman.plot(x='ZIP CODE', y='NUMBER OF PERSONS KILLED', alpha=0.1, kind="scatter", ax=ax, color="IndianRed")
ax.set_title("Motor Vehicle Collisions Manhattan")
```
