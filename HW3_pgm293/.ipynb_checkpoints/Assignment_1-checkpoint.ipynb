```python
from __future__ import print_function
import seaborn
import numpy as np
import pylab as pl
import pandas as pd
from scipy.optimize import curve_fit, minimize
from scipy.stats import norm

%pylab inline

np.random.seed(1000)
pop_mean = 100
```

```python
md = {}
#md['chisq'] = np.random.chisquare(pop_mean, size=100)
#print ("Chisq mean: %.2f, standard deviation: %.2f"%(md['chisq'].mean(), md['chisq'].std()))
#pl.hist(md['chisq'], bins = 20)
#pl.ylabel('N')
#pl.xlabel('x')
```

```python
mysize = np.random.randint(10,2000, size=100)
print (mysize,mysize.shape)
```

```python
df = pop_mean
md['chisq'] = {} 
for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size = n)
    
md['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str:
        md['chisq']['means'][nn] = md['chisq'][nn].mean()
        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi squared', fontsize=18)
        axchisq_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
```

```python
allmeans = list(md['chisq']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=20)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

```python
loc = pop_mean
md ['normal'] = {}
for n in mysize:
    md['normal'][n] = np.random.normal(loc, size = n)

md['normal']['means'] = {}
axnorm_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['normal']:
    if not type(nn) == str:
        md['normal']['means'][nn] = md['normal'][nn].mean()
        axnorm_mu_n.plot(nn,md['normal']['means'][nn],'o')
        axnorm_mu_n.set_xlabel('sample size', fontsize=18)
        axnorm_mu_n.set_ylabel('sample mean', fontsize=18)
        axnorm_mu_n.set_title('Normal', fontsize=18)
        axnorm_mu_n.plot([min(mysize),max(mysize)],[df,df],'k')
```

```python
allnormmeans = list(md['normal']['means'].values())
pl.figure(figsize=(10,10))
pl.hist(allnormmeans, bins=20)
pl.xlabel('sample mean', fontsize=18)
pl.ylabel('N', fontsize=18)
```

```python
lam = pop_mean
md['poisson'] = {}
for n in mysize:
    md['poisson'][n] = np.random.poisson(lam,size=n)

md['poisson']['means'] = {}
axpois_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['poisson']:
    if not type(nn) == str:
        md['poisson']['means'][nn] = md['poisson'][nn].mean()
        axpois_mu_n.plot(nn,md['poisson']['means'][nn],'o')
        axpois_mu_n.set_xlabel('sample size', fontsize=18)
        axpois_mu_n.set_ylabel('sample mean', fontsize=18)
        axpois_mu_n.set_title('Poisson', fontsize=18)
        axpois_mu_n.plot([min(mysize), max(mysize)], [lam, lam], 'k')
```
